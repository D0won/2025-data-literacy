```python
# 2017 ~ 2021 각 년도별 데이터를 딕셔너리 자료형에 저장
dict_data = {}
for year in range(2017, 2022) :
    data = pd.read_csv(f'./지역별 수온 및 염도/수온,염분_{year}.csv', encoding='cp949')
    if '조사년도' not in data.columns :
        data['조사년도'] = year
        data = data[['조사년도'] + [col for col in data.columns if col != '조사년도']]
    dict_data[year] = data
```
