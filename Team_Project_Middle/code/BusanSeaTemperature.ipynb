```python
import pandas as pd
```

```python
# 데이터 불러오기
temper = pd.read_csv('연안정지 관측자료.csv')
temper
```

```python
# 부산 키워드 있는 데이터만 뽑고 해파리 데이터와의 값 통일을 위하여 지역 칼럼 추가
temper = temper[temper['연안정점'].str.contains("부산")]
temper['지역'] = '부산'
```

```python
# 년도와 월별 데이터를 출력
temper['year'] = temper['관측일시'].astype(str).str[:4]
temper['month'] = temper['관측일시'].astype(str).str[4:6]
```

```python
# 수온 칼럼을 기준으로 Null 값이 들어있는 행 제거
temper.dropna(subset=['수온'], inplace=True)
```

```python
temper.to_csv('부산 수온.csv')
```
