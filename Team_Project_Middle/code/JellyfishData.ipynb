```python
import pandas as pd

jellyfishData = pd.read_csv('./해파리_모니터링_지리공간_데이터/해파리모니터링데이터_원본.csv')
jellyfishData
```

## 각 칼럼별 이름
| 칼럼명 | 데이터타입 | 칼럼 이름 |
| --- | --- | --- | 
|wtch_symd |	varchar |	관측시작일자 |
|wtch_eymd |	varchar	|관측종료일자 |
|jlyfs_nm	|varchar	|	해파리명 |
|txct_dgre_cd|	varchar	|독성정도코드 |
|ap_area_nm|	varchar	|출현지역명 |
|ap_dn_cd	|varchar	|출현밀도코드 | 
|jlyfs_aprt|	numeric		|해파리출현율 |
|wtch_iem_nm	|varchar		|관측항목명 |
|wtch_val	|numeric		|관측값 |

```python
# 관측항목명과 관측값은 null값밖에 없으므로 따로 뺌
jellyfishData = jellyfishData.drop(['WTCH_IEM_NM', 'WTCH_VAL'], axis= 1)
```

```python
# 널 개수 출력
null_counts = jellyfishData.isnull().sum()
print(null_counts)
```

```python
# 해파리 출현율 null값 처리 
jellyfishData.dropna(subset=['JLYFS_APRT'], inplace= True)
jellyfishData
```

```python
jellyfishData.to_csv('해파리 모니터링 데이터.csv', index = False, encoding='utf-8')
```
