### 데이터 내의 값(예: 남 vs. 여)을 따라 나누어 분석하기

```python
# 특정 조건인 데이터만 추출
data1[data1['sex']=='M']

# 참고
# data1[data1.sex=='M']
```

```python
# 특정 조건인 데이터만 추출
data1.query('sex=="M"')
```

```python
# 특정 조건인 데이터만 추출
data1[data1['sex']=='M'].describe()
```

```python
data1.groupby('sex').describe()
```

```python
# 히스토그램 겹쳐 그리기
sns.histplot(data1[data1.sex=='M'].weight, color='blue')
sns.histplot(data1[data1.sex=='F'].weight, color='red')
plt.show()
```

```python
# 상자그림
sns.boxplot(x = "sex", y  = "weight", data = data1, hue="sex")
#sns.boxplot(x = "sex", y  = "weight", data = data1, hue = "sex")
plt.show()
```

```python
# 바이올린그림
sns.violinplot(x = "sex", y  = "weight", data = data1)
plt.show()
```

```python
# 막대그래프
sns.barplot(x = "sex", y  = "weight", data = data1, hue = 'sex')
plt.show()
```
