## 문제5. while문의 활용
factorial값 N! = 1 * 2 * 3 * ... * (N-1) * N 로 정의된다.
N! 가 최초로 1000을 초과하게 되는 N과 N!값을 구해서 출력하시오.

```python
N = 1
num = 1
while True :
    if N > 1000 :
        break
    else :
        num *= N
        N += 1
print(N, num)
```

## 문제6. for 문의 활용
1~30 까지의 숫자 중 2, 3, 5의 배수만 빼고 출력하시오.

```python
for l in range(1, 31) :
    if l % 2 != 0 :
        if l % 3 != 0 :
            if l % 5 != 0 :
                print(l) 
```

## 문제7. 주민등록번호를 입력받아, 성별과 출생년도를 추출하시오.
input() 함수를 이용해 주민등록번호를 XXXXXX-XXXXXXX 형식으로 입력받을 것. 

문제 6-1) 주민등록번호에서 성별정보를 추출해 변수 gender = '남성' 또는 '여성'으로 저장하고, 출력하시오.

문제 6-2) 주민등록번호에서 출생년도를 계산해 변수 year 에 네자리수 값(OOOO)으로 저장하시오. (출생년도는 1901년~2022년 사이이다)

테스트 케이스로 다음을 사용해볼 것: 991111-1111111, 991111-2222222, 111111-3333333, 111111-444444

```python
id_num = input("주민등록번호를 입력하세요 (OOOOOO-XXXXXXX):")
# 문제 6-1
gender = 0
if id_num[7] == '1' or id_num[7] == '3' :
    gender = '남성'
else :
    gender = '여성'
print(gender)
# 문제 6-2
year = id_num[:2]
if int(year) < 23 :
    if id_num[8] == '1' :
        print('19' + year)
    else :
        print('20' + year)
else :
    print(19 + year)
```

## 문제8. 2~1000 사이의 소수 찾기
2~1000 범위에서 소수의 목록(또는 갯수)을 구하시오.

```python
import math
list2 = []
for i in range(2, 1001) :
    num = 0
    for j in range(2, int(math.sqrt(i))) :
        if i % j == 0 :
            num = 1
            
    if num == 0 :
        list2.append(i)
print(list2)
```
